```python
# Function to detect unnecessary characters in text
def contains_unnecessary_chars(text):
    pattern = r'[^\x00-\x7F]+'  
    if isinstance(text, str):
        return bool(re.search(pattern, text))
    return False

# Find columns with unnecessary characters
columns_with_unnecessary_chars = []

for column in df.columns:
    if df[column].apply(contains_unnecessary_chars).any():
        columns_with_unnecessary_chars.append(column)

print("Columns with unnecessary characters:")
print(columns_with_unnecessary_chars)
```

```python
# Columns where incorrect encodings are likely to occur
columns_to_check = ['Restaurant Name', 'City', 'Address', 'Locality', 'Locality Verbose', 'Cuisines', 'Currency']

# Finding unique values with unnecessary characters
unique_values_with_unnecessary_chars = {}

for column in columns_to_check:
    unique_values = df[column].unique()
    unwanted_chars_values = [val for val in unique_values if contains_unnecessary_chars(val)]
    unique_values_with_unnecessary_chars[column] = unwanted_chars_values

# Print the unique values with unnecessary characters
for column, values in unique_values_with_unnecessary_chars.items():
    print(f"Column: {column}")
    print(values)
    print()
```

## 1. Univariate Analysis (Top 10)
### Categorical Variables
### Frequency Distribution for Top 10 Categories

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of categorical columns
categorical_columns = ['Country Code', 'City', 'Locality', 'Cuisines', 'Currency', 
                       'Has Table booking', 'Has Online delivery', 'Is delivering now', 
                       'Switch to order menu', 'Rating color', 'Rating text']

# Function to plot the top 10 categories
def plot_top_10_categorical_distribution(column_name):
    plt.figure(figsize=(12, 6))
    top_10_categories = df[column_name].value_counts().nlargest(10).index
    top_10_data = df[df[column_name].isin(top_10_categories)]
    sns.countplot(data=top_10_data, y=column_name, order=top_10_categories)
    plt.title(f'Top 10 Categories Distribution for {column_name}')
    plt.xlabel('Count')
    plt.ylabel(column_name)
    plt.show()

# Plot for each categorical variable
for col in categorical_columns:
    plot_top_10_categorical_distribution(col)
```

### Numerical Variables
### Histogram and Boxplot for Top 10 Observations

```python
# List of numerical columns
numerical_columns = ['Average Cost for two', 'Price range', 'Aggregate rating', 'Votes']

# Function to plot histogram and boxplot for top 10 numerical values
def plot_top_10_numerical_distribution(column_name):
    top_10_values = df[column_name].nlargest(10)
    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    sns.histplot(top_10_values, kde=True)
    plt.title(f'Histogram of Top 10 {column_name}')
    
    plt.subplot(1, 2, 2)
    sns.boxplot(x=top_10_values)
    plt.title(f'Boxplot of Top 10 {column_name}')
    
    plt.show()

# Plot for each numerical variable
for col in numerical_columns:
    plot_top_10_numerical_distribution(col)
```

### a) Analyze the text reviews to identify the most common positive and negative keywords.

```python
# Define positive and negative ratings
positive_ratings = ['Excellent', 'Very Good', 'Good']
negative_ratings = ['Average', 'Poor', 'Terrible']

# Map ratings to positive and negative categories
df1['Sentiment'] = df1['Rating text'].apply(lambda x: 'Positive' if x in positive_ratings else 'Negative')

# Clean text reviews
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\W+', ' ', text)
    return text

# Clean 'Rating text' for analysis
df1['Cleaned_Rating_Text'] = df1['Rating text'].apply(clean_text)

# Separate positive and negative reviews
positive_reviews = df1[df1['Sentiment'] == 'Positive']['Cleaned_Rating_Text']
negative_reviews = df1[df1['Sentiment'] == 'Negative']['Cleaned_Rating_Text']

# Combine all reviews into a single string for each sentiment
positive_reviews_text = ' '.join(positive_reviews)
negative_reviews_text = ' '.join(negative_reviews)

# Define term mapping for positive and negative terms
positive_term_map = {
    'Excellent': ['excellent'],
    'Very Good': ['very good'],
    'Good': ['good']
}

negative_term_map = {
    'Average': ['average'],
    'Poor': ['poor'],
    'Terrible': ['terrible']
}

# Function to count multi-word phrases
def count_phrases(text, phrases):
    counts = Counter()
    for phrase in phrases:
        # Create a regex pattern for the phrase
        pattern = re.compile(r'\b' + re.escape(phrase.lower()) + r'\b', re.IGNORECASE)
        matches = pattern.findall(text)
        counts[phrase] += len(matches)
    return counts

# Count phrases in reviews
positive_phrase_counts = count_phrases(positive_reviews_text, positive_term_map.keys())
negative_phrase_counts = count_phrases(negative_reviews_text, negative_term_map.keys())

# Get most common keywords
most_common_positive = positive_phrase_counts.most_common()
most_common_negative = negative_phrase_counts.most_common()

print("Most common positive keywords:")
print(most_common_positive)

print("\nMost common negative keywords:")
print(most_common_negative)

# Plot bar charts
def plot_keywords(keyword_counts, title):
    words, counts = zip(*keyword_counts)
    plt.figure(figsize=(10, 5))
    plt.bar(words, counts, color='skyblue')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.show()

# Plot for positive and negative keywords
plot_keywords(most_common_positive, 'Most Common Positive Keywords')
plot_keywords(most_common_negative, 'Most Common Negative Keywords')

# Summary of the plots
positive_summary = "The most common positive keywords are 'Excellent', 'Very Good', and 'Good'."
negative_summary = "The most common negative keywords are 'Average', 'Poor', and 'Terrible'."

print("\nSummary of the plots:")
print(f"Positive Keywords: {positive_summary}")
print(f"Negative Keywords: {negative_summary}")
```

```python
from scipy.stats import chi2_contingency

# Create contingency tables
contingency_online_delivery = pd.crosstab(df['Price range'], df['Has Online delivery'])
contingency_table_booking = pd.crosstab(df['Price range'], df['Has Table booking'])

# Perform Chi-Square test
chi2_online_delivery, p_online_delivery, _, _ = chi2_contingency(contingency_online_delivery)
chi2_table_booking, p_table_booking, _, _ = chi2_contingency(contingency_table_booking)

print(f"Chi-Square Test for Online Delivery:\nChi2 Stat: {chi2_online_delivery}\nP-Value: {p_online_delivery}")
print(f"\nChi-Square Test for Table Booking:\nChi2 Stat: {chi2_table_booking}\nP-Value: {p_table_booking}")
```
